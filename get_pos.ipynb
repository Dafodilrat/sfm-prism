```python
import pandas as pd
import time
from scipy import integrate
import time
from tqdm.notebook import tqdm,trange
```

```python
acc_df= pd.read_csv('Data/imu_output_sync.csv')
```

```python
acc_df.head
```

```python
start_time=time.monotonic_ns()

vel_x=[]
vel_y=[]
vel_z=[]
vel_r=[]
vel_p=[]
vel_Y=[]
ax=[]
ay=[]
az=[]
gx=[];gy=[];gz=[]
current_time=[]
rows=[]

for row in trange(len(acc_df.index)):
  
  ax.append(acc_df['ax'][row])
  ay.append(acc_df['ay'][row])
  az.append(acc_df['az'][row])
  gx.append(acc_df['gx'][row])
  gy.append(acc_df['gy'][row])
  gz.append(acc_df['gz'][row])

  current_time.append(acc_df['time'][row]/1000000000)
  vel_x.append(integrate.trapezoid(ax,current_time))
  vel_y.append(integrate.trapezoid(ay,current_time))
  vel_z.append(integrate.trapezoid(az,current_time))
  vel_r.append(integrate.trapezoid(gx,current_time))
  vel_p.append(integrate.trapezoid(gy,current_time))
  vel_Y.append(integrate.trapezoid(gz,current_time))

  x=integrate.trapezoid(vel_x,current_time)
  y=integrate.trapezoid(vel_y,current_time)
  z=integrate.trapezoid(vel_z,current_time)
  r=integrate.trapezoid(vel_r,current_time)
  p=integrate.trapezoid(vel_p,current_time)
  Y=integrate.trapezoid(vel_Y,current_time)

  rows.append((acc_df['time'][row],x,y,z,r,p,Y))
```

```python
df = pd.DataFrame(
  rows,
  columns=['time','x', 'y', 'z', 'r', 'p', 'Y']
)

df.to_csv("Data/imu_pos_output.csv")
```
